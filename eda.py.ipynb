```python
import definitions as defs
import pandas as pd

training_df = pd.read_csv(defs.BASE_FOLDER +
                          defs.TRAINING_FILENAME)
```

```python
training_df.head()
```

```python
import cv2
max_width = 0
max_width_image = ""
min_width = 1000000000
min_width_image = ""
max_height = 0
max_height_image = ""
min_height = 1000000000
min_height_image = ""
for image_name in training_df['image_id'].values:
    img = cv2.imread(defs.BASE_FOLDER + 
                     defs.TRAINING_FOLDER + 
                     image_name, 0)
    height, width = img.shape
    if height > max_height:
        max_height = height
        max_height_image = image_name
    if height < min_height:
        min_height = height
        min_height_image = image_name
    if width > max_width:
        max_width = width
        max_width_image = image_name
    if width < min_width:
        min_width = width
        min_width_image = image_name

print("Max width: {}".format(max_width))
print(max_width_image)
print("Min width: {}".format(min_width))
print(min_width_image)
print("Max height: {}".format(max_height))
print(max_height_image)
print("Min height: {}".format(min_height))
print(min_height_image)
```
